```python
# works for result table file generated from MM analyzer ImageJ plugin, 
# even more than one repeated row (multiple gfp or rfp in the same cell)
ts =[]; 
lengths =[];
names = [];

divTime=[];
divLength = [];
divName = [];

# only consider mother cell name started by "A"
def isHeadCell(s):
    if s.find('A')<0:
        return False
    return s.find('T') < 0

with open('./haotian position 1/9.txt','r') as f:
    for line in f:      
        #text separation parsed by tab
        s = line.split('\t')
        if s[0] == 'cell_label':            
            indexName  = s.index('name');
            indexTime = s.index('timeIndex');
            indexLen = s.index('length');
            continue;
        if isHeadCell(s[indexName]):
            names.append(s[indexName])
            lengths.append(int(s[indexLen]))
            ts.append(int(s[indexTime ]))
    
    for j,val in enumerate(names):
        if j==len(names) - 1:
            break;
        if len(names[j]) < len(names[j+1]):
            divTime.append(ts[j]);
            divName.append(names[j]);
            divLength.append(lengths[j])
        
f.close()

if(len(divLength)<2):
    print("Warning ! Less than two division, the plot will be abandoned")
    %matplotlib inline  
    import matplotlib.pyplot as plt
    plt.figure(1,figsize=(10,5))
    plt.subplot(211)
    plt.plot(ts, lengths,'b')
    plt.plot(divTime, divLength,'r*')
```

```python
interval = 5;
l1 = divTime[1:len(divTime)];
l2 = divTime[0:len(divTime)-1];
l_sub = []
for ind,val in enumerate(l1) :
    l_sub.append((l1[ind] - l2[ind])*interval);
l_sub.insert(0,l_sub[0])
```

```python
%matplotlib inline  
import matplotlib.pyplot as plt
import numpy as np
plt.figure(1,figsize=(10,5))
plt.subplot(211)
plt.plot(ts, lengths,'b')
plt.plot(divTime,divLength,'*r')
plt.xticks(np.arange(min(ts),max(ts),2.0))
plt.ylabel('cell length')
plt.xlabel('time-minutes')
plt.axis([0, max(ts)+5, min(lengths)-5, max(lengths)+5])

plt.subplots_adjust(hspace = 0.5)

plt.subplot(212)
plt.plot(divTime[0:len(divTime)],l_sub,'g')
plt.ylabel('division time')
plt.xlabel('time-minutes')
plt.axis([0, max(ts)+5, min(l_sub)-5, max(l_sub)+5])
plt.plot(divTime,l_sub,'*r')
plt.yticks(np.arange(min(l_sub),max(l_sub)+5,5.0))
#plt.axis([0, max(ts)+5, min(lengths)-5, max(lengths)+5])
plt.show()
```
